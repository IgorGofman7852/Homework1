```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
function = lambda x: 3 * x ** 2
df = lambda x: 6 * x
x =  np.arange (-5, 6)
plt.plot(x, function(x))
```

```python
x0 = 5
L = 0.1
res_x = []
res_function = []
x1 = x0 - L * df(x0)
```

```python
while np.linalg.norm(x0 - x1) > 0.00000001:
    res_x.append(x0)
    res_function.append(function(x0))
    x0 = x1
    x1 = x0 - L * df(x0)
```

```python
plt.plot(res_x, res_function)
plt.plot(x, function(x))
```

```python
def function_2(x, y):
    f = 3 * x ** 2 + 5 * y ** 2
    return(f)
```

```python
def df_2(x, y):
    df_x = 6 * x
    df_y = 10 * y
    return(df_x + df_y)
```

```python
x0 = 5
y0 = 10
L = 0.1
res_x = []
res_y = []
res_function = []
x1 = x0 - L * df_2(x0, y0)
```
